```python
df.drop(['Benefits','TotalPayBenefits','Notes','Status'],axis=1,inplace=True)
#I have removed all null columns and repeated columns also
```

```python
df
```

```python
df=df[~df['EmployeeName'].str.contains('Not provided')]
df
#I have removed the rows not contain any name ('Not provided')
```

```python
df.info()
#after data cleaning 
```

```python
df['TotalPay'].describe()
```

```python
range = df['TotalPay'].max() - df['TotalPay'].min()

print(f"The range of salary is {range}.")
```

```python
median = np.median(df['TotalPay'])

print(f"The median of the array is {median}.")
```

```python
mode = df['TotalPay'].mode()[0]

print(f"The mode of column is {mode}.")
```

```python
df['TotalPay'].plot.hist(bins=10, color='green')
```

```python
df.groupby(df['Year'])[['TotalPay','BasePay','OvertimePay','OtherPay']].mean()
```
